```python
HOFdata = pd.read_csv('./eyecolorgenderdata.csv') 
HOFdata.head()
```

```python
def cin_subsample_dataframe(df_input: pd.DataFrame, cin: int) -> pd.DataFrame:
    last_two_digits = cin%100
    selection_value = last_two_digits//4
    selection_vector = np.arange(0, len(df_input),selection_value)
    return df_input.iloc[selection_vector]
```

```python
hof_subsample=cin_subsample_dataframe(HOFdata,401975311)
hof_subsample.head()
```

```python
Name_Prob = {}

eyecolor = hof_subsample[hof_subsample.gender == 'male']['eyecolor']
dum = hof_subsample[hof_subsample.gender == 'male']['eyecolor'].unique()

for i in dum:
  Name_Prob[i] = len(eyecolor[eyecolor == i]) / len(eyecolor)
```
